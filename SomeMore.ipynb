```python
'''Q - Find the top 5 players with the most four in IPL 2016.'''
rungroup = data.groupby('batsman_runs')
four = rungroup.get_group(4)['batsman'].value_counts()
print(four.head(5))  # Final Answer
```

```python
'''Virat Kohli's ke sabse jyada runs kis team ke khilaf hai?'''
vk = data[data['batsman'] == 'V Kohli']
vk.head(2)
vk_runs = vk.groupby('bowling_team')['batsman_runs'].sum()
vk_runs.sort_values(ascending=False).head(5)
```

```python
def team_catny(player_name):
    player_data = data[data['batsman'] == player_name]
    player_runs = player_data.groupby('bowling_team')['batsman_runs'].sum()
    return player_runs.sort_values(ascending=False).head(5)

team_catny('V Kohli')
```

```python
# Find the most destructive batsman in IPL 2016.
# Minimum 200 balls should be played by the batsman.
# Strite Rate = (Total Runs Scored / Total Balls Played) * 100
data.head(3)
```

```python
last_overs = data[data['over'] > 15] # Only considering the last 5 overs of the innings
selectednames = last_overs['batsman'].value_counts()[last_overs['batsman'].value_counts() > 150].index
print("Selected Batsman Names:", selectednames)
```

```python
last_overs = last_overs[last_overs['batsman'].isin(selectednames)]
last_overs.head(3)
```

```python
grp_batsman = last_overs.groupby('batsman')
runs = grp_batsman['batsman_runs'].sum()
runs
```

```python
balls = grp_batsman['ball'].count()
balls
```

```python
top = runs / balls * 100
top.sort_values(ascending=False).head(5)
```

```python
delhivery = pd.read_csv('deliveries.csv')
matches = pd.read_csv('matches.csv')
```

```python
# Merging the two dataframes on 'id' and 'match_id'
merged_data = delhivery.merge(matches, left_on='match_id', right_on='id')
```

```python
merged_data.head(3)

```

```python
merged_data.info()
```

```python
runs_data = merged_data.groupby(['season', 'batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
runs_data.head(10)
runs_data.drop_duplicates(subset=['season']).sort_values(by='season', ascending=False).reset_index(drop=True)
```
